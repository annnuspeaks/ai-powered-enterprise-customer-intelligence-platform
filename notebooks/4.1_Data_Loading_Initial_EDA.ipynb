```python
customers_clean = customers.copy()
geolocation_clean = geolocation.copy()
order_items_clean = order_items.copy()
order_payments_clean = order_payments.copy()
order_reviews_clean = order_reviews.copy()
orders_clean = orders.copy()
products_clean = products.copy()
sellers_clean = sellers.copy()
category_translation_clean = category_translation.copy()
```

```python
date_columns = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in date_columns:
    if col in orders_clean.columns:
        orders_clean[col] = pd.to_datetime(
            orders_clean[col],
            errors="coerce"
        )
```

```python
clean_datasets = {
    "Customers": customers_clean,
    "Geolocation": geolocation_clean,
    "Order Items": order_items_clean,
    "Order Payments": order_payments_clean,
    "Order Reviews": order_reviews_clean,
    "Orders": orders_clean,
    "Products": products_clean,
    "Sellers": sellers_clean,
    "Category Translation": category_translation_clean
}

for name, df in clean_datasets.items():
    before = len(df)
    df.drop_duplicates(inplace=True)
    after = len(df)

    print(f"{name:<25} Removed: {before-after}")
```

```python
master_df = customer_orders.copy()

print(master_df.shape)
master_df.head()
```

```python
master_df["total_price"] = (
    master_df["price"] +
    master_df["freight_value"]
)
```

```python
reference_date = master_df["order_purchase_timestamp"].max()

print(reference_date)
```

```python
customer_features = (
    master_df
    .groupby("customer_unique_id")
    .agg(
        Recency=(
            "order_purchase_timestamp",
            lambda x: (reference_date - x.max()).days
        ),

        Frequency=(
            "order_id",
            "nunique"
        ),

        Monetary=(
            "total_price",
            "sum"
        ),

        Total_Orders=(
            "order_id",
            "nunique"
        ),

        Total_Spent=(
            "total_price",
            "sum"
        ),

        Average_Order_Value=(
            "total_price",
            "mean"
        ),

        Average_Review_Score=(
            "review_score",
            "mean"
        )
    )
    .reset_index()
)
```

```python
delivery = (
    master_df
    .groupby("customer_unique_id")
    .apply(
        lambda x: (
            x["order_delivered_customer_date"] -
            x["order_purchase_timestamp"]
        ).dt.days.mean()
    )
    .reset_index(name="Average_Delivery_Days")
)

customer_features = customer_features.merge(
    delivery,
    on="customer_unique_id",
    how="left"
)
```

```python
ml_dataset = customer_features.copy()

ml_dataset.head()
```

```python
ml_dataset["Churn"] = (
    ml_dataset["Recency"] > 180
).astype(int)
```

```python
def customer_segment(row):

    if row["Monetary"] >= 1000:
        return "High Value"

    elif row["Frequency"] >= 5:
        return "Loyal"

    elif row["Recency"] <= 60:
        return "Active"

    else:
        return "Regular"


ml_dataset["Customer_Segment"] = ml_dataset.apply(
    customer_segment,
    axis=1
)
```

```python
ml_dataset["Predicted_CLV"] = (
    ml_dataset["Total_Spent"] *
    ml_dataset["Frequency"]
)
```

```python
ml_dataset["Purchase_Propensity"] = (
    ml_dataset["Recency"] <= 90
).astype(int)
```
